Удаление пропусков и демонстрация результата удаления

```python
df = df.dropna(subset=['salary'])
```

```python
print(df.isna().sum())
```

Подсчет количества и удаление явных дубликатов

```python
print(f"Найдено {df.duplicated().sum()} дубликата(ов)")
df=df.drop_duplicates()
df.info()
```

Поиск неявных дубликатов в столбцах `job_title`, `employee_residence`,  `company_location`, `company_size`


```python
for col in df.columns:
  if col in ['job_title', 'employee_residence', 'company_location', 'company_size']:
    print(df[col].unique())
```

Устранение неявных дубликатов (замена написаний терминов)

```python
df['job_title']=df['job_title'].replace('Data SCIENTIST', 'Data Scientist')
df['job_title']=df['job_title'].replace('DataScientist', 'Data Scientist')
df['job_title']=df['job_title'].replace('Data AnalyticsManager', 'Data Analytics Manager')
df['company_size']=df['company_size'].replace('Large', 'L')
```

Конвертация в числовой формат столбца `work_year`

```python
df['work_year'] = df['work_year'].astype(int)
df.info()
```

Группировка по типу трудоустройства с сортировкой по количеству компаний

```python
df2=df.groupby(['employment_type', 'company_location'])['company_location'].count().sort_values()
df2=df2.to_frame() #преобразуем series в dataframe
df2=df2.rename(columns={'company_location': ' count'}) #переименовываем колонну
df2.head(100)
```

Группировка `employment_type`  и кпо каждой локации

```python
df.groupby(['employment_type', 'company_location'])['company_location'].count()
```

Построение сводной таблицы для нахождения средней зарплаты с округлением до 3 знаков после запятой и сортировкой по убыванию

```python
pd.pivot_table(
    df,
    index='work_year',
    values='salary_in_usd',
    aggfunc='mean'
).rename(columns={'salary_in_usd': 'зарплата'}).sort_values(ascending=False, by='зарплата').round(3)
```

```python
pd.pivot_table(
    df,
    index='work_year',
    columns='employment_type',
    values='salary_in_usd',
    aggfunc='median'
).sort_values(ascending=False, by='work_year')
```
